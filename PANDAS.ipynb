# Group by 

```python
#Group by is going to group together the values in a column and display all in the same row and allows you to perform those aggrerate functions on those groupings
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\ANALYSIS\CLEAN DATA\Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame
```

```python
group_by_frame.mean()
```

```python
df
```

```python
df.groupby('Base Flavor').mean()
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'],'Texture Rating': ['mean','max','count','sum']})
```

```python
df.groupby('Base Flavor').describe()
```
